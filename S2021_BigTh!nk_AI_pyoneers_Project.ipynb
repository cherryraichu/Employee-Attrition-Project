# The Problem

Rivers Cuomo (legendary frontman of the rock band [Weezer](https://www.youtube.com/watch?v=ENXvZ9YRjbo&list=PL-4cSyrIouqQ5nnYRGtXUg7nM0myc83XT)) is the CEO of **Water4Everyone**, a non-profit that sells low-cost water purification systems to governments of developing nations. Access to clean water is serious issue: about 2.2 billion people in the world do not have safely managed water drinking services. For the last 5 years, Water4Everyone has been able to provide over 20 countries and 20 million people with access to sanitized water. 

The Head of HR (Human Resources) informs Cuomo that **Water4Everyone's** annual employee attrition rate has increased drastically over the last two years. Unfortunately, The HR department is not sure why this is happening.

Consequentially, the Finance department also informs Cuomo that since recruiting/training new employees costs a significant amount of money, if this trend continues, the company may go bankrupt by the end of next year. Cuomo decides to work with **BigTh!nk AI Consulting™** to solve this issue.

### Goal

Given a dataset of **Water4Everyone's** current employees, the **.pyoneers** division's job is to process this data and build a machine learning model that will predict whether a certain employee will attrite (leave) the company. This will allow **Water4Everyone's** HR department to identify employees predicted to attrite and negotiate terms that would motivate them to stay.
